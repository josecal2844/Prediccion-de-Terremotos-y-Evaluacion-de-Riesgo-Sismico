# Fase 1 — Exploración y preprocesamiento de datos

Carga de los datos locales (HDF5 y CSV de STEAD), filtrado y preparación para su uso en los modelos de aprendizaje profundo.

```python
import pandas as pd
import numpy as np
import h5py
import matplotlib.pyplot as plt
import os
import sys

sys.path.append(os.path.abspath(""))
```

```python
from utils.data_loaders import STEADDataset
import torch

CSV_PATH = os.path.join("..", "STEAD", "features.csv")
HDF5_PATH = os.path.join("..", "STEAD", "waveforms.hdf5")

if os.path.exists(CSV_PATH) and os.path.exists(HDF5_PATH):
    print("Archivos STEAD encontrados. Cargando...")
    df = pd.read_csv(CSV_PATH)
    df_sample = df.sample(100, random_state = 42)

    dataset = STEADDataset(df_sample, HDF5_PATH)
    with dataset:
        onda, nombre = dataset[0]
        print(f"Onda cargada: {nombre} | Dimensiones: {onda.shape} (Canales, Longitud)")

        canales = ["E", "N", "Z"]
        fig, axs = plt.subplots(3, 1, figsize = (10, 6), sharex = True)
        for i in range(3):
            axs[i].plot(onda[i].numpy(), color = "C" + str(i))
            axs[i].set_ylabel(f"Canal {canales[i]}")
        axs[2].set_xlabel("Muestras temporales (100 Hz)")
        plt.suptitle(f"Sismograma de la forma de onda: {nombre}")
        plt.tight_layout()
        plt.show()
else:
    print("Los archivos STEAD no se han encontrado en las rutas especificadas.")
    print("Ajusta 'CSV_PATH' y 'HDF5_PATH' a donde tengas los archivos descargados.")
```
